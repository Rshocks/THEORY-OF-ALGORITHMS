### Exercise 1 $f(x) = x^2 + 1$
***

```python
# Define the function
def f(x):
    return x**2 + 1

# Create an array of x values
x = np.linspace(-10, 10, 1000)

# Calculate the corresponding y values
y = f(x)

# Plot the function
plt.plot(x, y)

# Set the axis labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x^2 + 1')

# Display the plot
plt.show()
```

### Exercise 2 $g(x) = 3x^6 + x^4 + 1$
***

```python
def g(x):
    return 3*x**6 + x**4 + 1

x = np.linspace(-2, 2, 1000)

y = g(x)

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Plot of g(x) = 3x^6 + x^4 + 1')

plt.show()
```

### Exercise 3 $ x^2 $ and $2^x$ on the same plot
***

```python
# create a range of x-values
x = np.linspace(0, 5, 100)

# calculate y-values for x^2 and 2x
y1 = x**2
y2 = 2**x

# plot the functions on the same graph
plt.plot(x, y1, label='x^2')
plt.plot(x, y2, label='2x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x^2 and 2x')
plt.legend()

plt.show()
```
