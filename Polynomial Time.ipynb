# Polynomial Time
***

Polynomial time is a term used to describe how long it takes for an algorithm or computer program to complete. It implies that the time required will increase no more swiftly than a polynomial function of the problem size. Simply, it indicates that as the magnitude of the problem grows, the time required will grow, but not quickly.

In the sense that we have 100 operations with quadratic O(n*n) (where n is 10) and 100 operations with polynomial O(n2) (where n is 10), so you could say that a polynomial time complexity method is functionally a quadratic algorithm.

When the polynomial equation's exponent (2) is raised, the difference now becomes apparent (where as with quadratic it will always be a squared number). For instance, O(n3) is also described as polynomial.[5]

## Big O Notation
***
Big O is a way of describing time complexity is is normally written in the form O(f(n)) where n is the input size and O is the order of magnitude, such as to polynomial time complexity. It is a important algorithm is establishing preformance and how input size will increase over time.

For Example if a problem is O(n) means that the maximum amount of time it will take to run increases linearly with the input size. 

An algorithm with a time complexity of O(n^2) has polynomial time complexity, because the maximum amount of time it takes to run increases quadratically with the input size [15] .

### Big O Notation & polynomial time complexity examples
***
The below example should have a output of a empty list because no pair in the input list add to 10, this means there is a polynomial time complexity as nested loops are iterated over all possible pairs of numbers in the input list

```python
def find_pairs_sum(numbers, target):
    # This function finds all pairs of numbers in a list that add up to a target value
    pairs = set()
    for i in range(len(numbers)):
        for j in range(i+1, len(numbers)):
            if numbers[i] + numbers[j] == target:
                pairs.add((min(numbers[i], numbers[j]), max(numbers[i], numbers[j])))
    return list(pairs)

numbers = [1, 2, 3, 4, 5]
target = 10

pairs = find_pairs_sum(numbers, target)
print(pairs)
```

The below example is a graph of big O in polynomial time, it takes a int n as input, then returns the sum of a squares of all integers from 1 to n. It is a polynomia

```python
def polynomial_function(n):
    # This function takes an integer n as input and returns the sum of the squares of all integers from 1 to n
    sum = 0
    for i in range(1, n+1):
        for j in range(1, n+1):
            sum += i*j
    return sum

# Test the function with different input sizes and measure the time it takes to run
input_sizes = [10, 50, 100, 200, 500, 1000]
runtimes = []
for n in input_sizes:
    start_time = time.time()
    polynomial_function(n)
    end_time = time.time()
    runtimes.append(end_time - start_time)

# Plot the results
plt.plot(input_sizes, runtimes)
plt.xlabel('Input Size')
plt.ylabel('Time (Seconds)')
plt.title('Time Complexity of Polynomial Function')
plt.show()
```

### Dynamic programming
***
Dynamic programming algorithms are often designed to solve optimization problems, and many of these algorithms have polynomial time complexity.

The time complexity of dynamic programming algorithms is often represented using big O notation, which provides an upper bound on the running time of the algorithm.

Example: Dynamic programming algorithm can be used for finding the longest common subsequence of two sequences which have a time complexity of O(mn), where m and n are the lengths of the two sequences. It runs on polynomial time because the run time increases no greater than quadratically with the length of the input. [14]

```python
def fibonacci(n):
    if n < 0:
        return None
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    fib_table = [0] * (n+1)
    fib_table[1] = 1
    
    for i in range(2, n+1):
        fib_table[i] = fib_table[i-1] + fib_table[i-2]
        
    return fib_table[n]
```

### Sorting lists which run in polynomial time
***

- Quicksort works as a big array which then splits the data from the first big array into a number of smaller arrays. Quicksort works as two iterations sorting the aforementioned array, the next is putting the array back together. Quicksorts complexity is O(n*logn), quicksort normally splits the smaller elements to the left and larger elements to the right [6].


- Mergesort works in a familiar way as quicksort works primarly using divide and conquer of which is a algorithm which breaks down the problem into further sub components. In mergesort it divides the problem into two components both are equal, this process is repeated until it can't be done anymore. Next, we join the two components into two-element lists, until we reach the sorted list, the sorted two-element pairs are combined with the four-element lists until the lists are sorted [7].


- Heapsort gets rid of elements one by one from a heaped list, therein sorting them into a sorted list. A heap itself is a binary tree, in which each node in the tree will have no more than two children nodes, all the binary tree levels should be filled save for the last level [8].

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        left = []
        right = []
        for i in range(1, len(arr)):
            if arr[i] < pivot:
                left.append(arr[i])
            else:
                right.append(arr[i])
        return quicksort(left) + [pivot] + quicksort(right)
    
arr = [3, 7, 2, 9, 1, 6, 8, 4, 5]

sorted_arr = quicksort(arr)

print(sorted_arr)
```

```python
# mergesort example
def mergesort(arr):
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    
    left = mergesort(left)
    right = mergesort(right)
    
    return merge(left, right)

def merge(left, right):
    result = []
    i = 0
    j = 0
    
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    result += left[i:]
    result += right[j:]
    
    return result

arr = [5, 3, 8, 4, 2, 7, 1, 6]

sorted_arr = mergesort(arr)

print(sorted_arr)
```

```python
# heapsort example
def heapsort(arr):
    n = len(arr)
    
    # Build a max-heap from the array
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
    
    # Extract elements from the heap one by one
    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, i, 0)
    
    return arr

def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2
    
    # Check if left child is larger than root
    if left < n and arr[left] > arr[largest]:
        largest = left
    
    # Check if right child is larger than root
    if right < n and arr[right] > arr[largest]:
        largest = right
    
    # If root is not the largest, swap with the largest child and continue heapifying
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

arr = [5, 3, 8, 4, 2, 7, 1, 6]

sorted_arr = heapsort(arr)

print(sorted_arr)
```

### Dijkstra's algorithm
***
Finding the shortest path between two nodes in a weighted graph using Dijkstra's algorithm, which has a worst-case time complexity of O((|V| + |E|) log |V|), where |V| is the number of vertices and |E| is the number of edges in the graph.


The method uses a priority queue of nodes to explore, where the priority is the distance to the node. 

The start node is added to the queue after all nodes except the first node, whose distance is zero, their distances are then initialized to infinity. Then each time a shorter path is discovered, it updates the distances of its neighbors, adds them to the queue in addition to repeatedly selecting the node with the shortest distance from the queue.


Once the end node has been reached, the algorithm reconstructs the shortest path by going backwards from the end node to the start node using a dictionary of predecessors that records the prior node for each node that has been visited. The algorithm returns None if there is no path between the start and finish nodes [9] [10].

```python
def dijkstra(graph, start, end):
    # Initialize the distance of all nodes to infinity except the start node which is 0
    distances = {node: float('inf') for node in graph}
    distances[start] = 0

    # Initialize the priority queue with the start node and its distance
    queue = [(0, start)]

    # Initialize the predecessors dictionary to keep track of the shortest path
    predecessors = {}

    while queue:
        # Get the node with the smallest distance
        current_distance, current_node = heapq.heappop(queue)

        # If weve reached the end node return the shortest distance and the shortest path
        if current_node == end:
            path = []
            while current_node in predecessors:
                path.append(current_node)
                current_node = predecessors[current_node]
            path.append(start)
            path.reverse()
            return (distances[end], path)

        # For each neighbor of the current node calculate the new distance and update the distance if its shorter
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                predecessors[neighbor] = current_node
                heapq.heappush(queue, (distance, neighbor))

    # If exhausted all possible paths and havent reached the end node there is no path
    return (float('inf'), None)
```

### Matrix multiplication
***
The problem of multiplying two matrices to obtain a new matrix is known as matrix multiplication. 

The product of two matrices A and B of dimension n x n is a new matrix C of dimension n x n, with each entry C[i][j] being the sum of the products of the corresponding elements in row I of matrix A and column j of matrix B.

The time complexity of the naive algorithm for matrix multiplication is O(n3), which can be very slow for large matrices [11]. 

```python
def matrix_mult(A, B):
    n = len(A)
    C = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C
```

### Matrix multiplication using Coppersmith–Winograd algorithm
***
The Coppersmith–Winograd algorithm is a matrix multiplication algorithm, first coined in 1987.

The algorithm achieves a time complexity of O(n^2.376), which is an improvement over the naive O(n^3) algorithm for multiplying two n×n matrices. 

The algorithm is based on the idea of recursively dividing the input matrices into smaller submatrices, and then using clever methods for combining the results of these submatrix multiplications.[12][13]

```python
def coppersmith_winograd(A, B):
    n = A.shape[0]
    if n == 1:
        return A[0, 0] * B[0, 0]
    
    if n % 2 == 1:
        A = np.pad(A, [(0, 1), (0, 1)], mode='constant')
        B = np.pad(B, [(0, 1), (0, 1)], mode='constant')
    
    m = n // 2
    A11 = A[:m, :m]
    A12 = A[:m, m:]
    A21 = A[m:, :m]
    A22 = A[m:, m:]
    B11 = B[:m, :m]
    B12 = B[:m, m:]
    B21 = B[m:, :m]
    B22 = B[m:, m:]
    
    S1 = B12 - B22
    S2 = A11 + A12
    S3 = A21 + A22
    S4 = B21 - B11
    S5 = A11 + A22
    S6 = B11 + B22
    S7 = A12 - A22
    S8 = B21 + B22
    S9 = A11 - A21
    S10 = B11 + B12
    
    P1 = coppersmith_winograd(A11, S1)
    P2 = coppersmith_winograd(S2, B22)
    P3 = coppersmith_winograd(S3, B11)
    P4 = coppersmith_winograd(A22, S4)
    P5 = coppersmith_winograd(S5, S6)
    P6 = coppersmith_winograd(S7, S8)
    P7 = coppersmith_winograd(S9, S10)
    
    C11 = P5 + P4 - P2 + P6
    C12 = P1 + P2
    C21 = P3 + P4
    C22 = P5 + P1 - P3 - P7
    
    C = np.vstack((np.hstack((C11, C12)), np.hstack((C21, C22))))
    
    if n % 2 == 1:
        C = C[:-1, :-1]
    
    return C

# Example usage
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = coppersmith_winograd(A, B)
print(C[0, 0])  # Prints 19
```

### Graph Isomorphism
***
Graph Isomorphism is a computer science problem to assert if two graphs are isomorphic(if there is a one to one correspondence between vertices) [16].

Graph Isomorphism was not known to be solvable in polynomial time, it is known the be not solvable because the nodes on the graphs grow very very quick.

In 2015 László Babai produced a new algorithm which could run in a somewhat polynomial time complexity, which was an improvement over the best known algorithm for this issue before, exponential time [17].

The below example in python is just a simple set of if statements which attempt to solve the graph isomorphism problem in polynomial time by simply comparing nodes and edges in the two sets, the below algorithm may only be correct in simple cases not complex ones.

```python
# Define two example graphs
G1 = nx.Graph()
G1.add_edges_from([(1,2), (1,3), (2,3)])

G2 = nx.Graph()
G2.add_edges_from([(4,5), (5,6), (4,6)])
```

```python
def is_isomorphic(G1, G2):
    # Check if the number of nodes and edges are the same
    if G1.number_of_nodes() != G2.number_of_nodes() or G1.number_of_edges() != G2.number_of_edges():
        return False
    # Check if the node and edge sets are the same
    if set(G1.nodes()) != set(G2.nodes()) or set(G1.edges()) != set(G2.edges()):
        return False
    # Check if the degree sequences are the same
    if sorted(G1.degree()) != sorted(G2.degree()):
        return False
    # If all tests pass, the graphs are isomorphic
    return True
```

### Traveling Salesman Problem
***
The traveling salesman problem(TSP) is a issue where is a salesperson has a checklist of locations to visit while he is selling his wears, what is the quickest route back to his starting point ? [18]

TSP is not known to have a polynomial time algorithm, if TSP can be solved in polynomial time it can solve many problems for example, optimization problems in computer science [19].

The main reason TSP is difficult to solve is that there is no quick solution, and the complexity increases as more locations are visited.

The most common algorithm to solve this problem, is brute force as shown below in python.

```python
def tsp(cities):
    """
    Given a set of cities and their distances to each other,
    find the shortest possible route that visits each city exactly once
    and returns to the starting city.
    """
    shortest_route = None
    shortest_distance = float('inf')
    
    for route in itertools.permutations(cities):
        distance = 0
        
        for i in range(len(route)):
            # Add the distance from the current city to the next city in the route
            distance += cities[route[i]][route[(i+1)%len(route)]]
            
        # Update the shortest distance and route found so far
        if distance < shortest_distance:
            shortest_distance = distance
            shortest_route = route
    
    if shortest_route:
        print(f"Shortest route: {' -> '.join(shortest_route)} -> {shortest_route[0]}")
        print(f"Shortest distance: {shortest_distance}")
    else:
        print("No solution found.")
```

```python
cities = {'A': {'B': 2, 'C': 3, 'D': 1},
          'B': {'A': 2, 'C': 4, 'D': 2},
          'C': {'A': 3, 'B': 4, 'D': 5},
          'D': {'A': 1, 'B': 2, 'C': 5}}

tsp(cities)
```
