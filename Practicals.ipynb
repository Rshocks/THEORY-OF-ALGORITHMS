## Turing Machine
***
A Turing machine is a hypothetical computer that uses a tape with symbols, a read-write head, and a set of rules to perform computations. It can simulate any algorithm and is a theoretical basis for modern computers

[Turing Machine on YouTube](https://www.youtube.com/watch?v=E3keLeMwfHY)

[Turing Machine on Explained](https://www.futurelearn.com/info/courses/how-computers-work/0/steps/49259)

```python
# The huring machine tape
tape = " "

# Pos on the tape
pos = 0

# show tape and pos
print(f'{tape}, {pos}')

# moving pos left of tape
def move_L():
    global tape, pos
    if pos > 0:
        pos = pos - 1
    else:
        tape = ' ' + tape
```

```python
move_L()

print(f'{tape}, {pos}')
```

```python
# moving pos right of tape
def move_R():
    global tape, pos
    if pos >= len(tape):
        tape = ' ' + tape
    pos = pos + 1
```

```python
move_R()

print(f'{tape}, {pos}')
```

```python
# A Turing Machine State Table Example.
table = {
  ('s', '0'): ('0', 'R', 's'),
  ('s', '1'): ('1', 'R', 't'),
  ('s', '_'): ('_', 'R', 'a'),
  ('t', '0'): ('0', 'R', 't'),
  ('t', '1'): ('1', 'R', 's'),
  ('t', '_'): ('_', 'R', 'f'),
}
```

```python
# Start state.
state = 's'

# Show state.
state

# Symbol in current cell on tape.
tape[pos]

#table[(state, tape[pos])]
```

```python
def run_turing_machine(table, tape, state):
    # Strings are immutable.
    tape = list(tape)
    
    # Check for empty string.
    tape.append('_')
    
    # Start at position 0.
    pos = 0
    
    # Show tape and state.
    output = ''.join(tape)
    output = output[:pos] + state + output[pos:]
    print(output)
    
    # Keep looping unless we enter qa or qf.
    while state not in {'a', 'f'}:
        
        # Look up the table for current state/symbol.
        overwrite, move, nextstate = table[(state, tape[pos])]
        
        # Overwrite the current cell.
        tape[pos] = overwrite
        
        # Move in the correct direction.
        if move == 'L':
            if pos > 0:
                pos -= 1
            else:
                tape.insert(0, '_')
        else:
            pos += 1
            if pos >= len(tape):
                tape.append('_')
                
        # Change to next state.
        state = nextstate
        
        # Show tape and state.
        output = ''.join(tape)
        output = output[:pos] + state + output[pos:]
        print(output)
        
    # Check the state.
    return state == 'a'
```

```python
# Try an example.
run_turing_machine(table, '01111', 's')
```

## String Matching Indices Lab 22/02/23 
***
[String Matching Indices](https://seqan.readthedocs.io/en/master/Tutorial/Algorithms/PatternMatching/IndexedPatternMatching.html)

Given two lists of strings L1 and L2 of the same length, return a list L of indices of L1 and L2 such that L1[i] and L2[j] have the same characters for all positions in the range [0, len(L1)), and such that |L| <= 10.

```python
def correspondence(L1, L2):
    L = []
    for i in range(len(L1)):
        for j in range(len(L2)):
            if i not in L and j not in L and L1[i] == L2[j]:
                L.append(i)
                L.append(j)
                if len(L) >= 10:
                    return L
    return L

L1 = ['010', '001', '111', '000']
L2 = ['010', '101', '011', '000']
L = correspondence(L1, L2)
print(L)
```

## Lab 01/03/2023 Boolean Satisfiability Problem
***
The issue of establishing whether a Boolean formula is satisfiable or unsatisfiable is known as Boolean Satisfiability(SAT).The formula is said to be satisfiable if the Boolean variables can be given values that result in the formula being true.


[Wiki on Boolean Satisfiability Problem](https://en.wikipedia.org/wiki/Boolean_satisfiability_problem)

[Engati on Boolean Satisfiability Problem](https://www.engati.com/glossary/boolean-satisfiability-problem)


Write a python script to find all settings of p,q and r that make these true.


1) $(p 갿 춻q) 갾 (춻p 갿 q 갿 r) 갾 춻p$ Consisting of three clauses, each of which contains two or three variables. The goal is to find a truth assignment to the variables p, q, and r that satisfies all three clauses simultaneously.


2) $p 갿 q 갾 (춻p 갿 춻q)$  It contains two clauses, each of which contains two variables. The goal is to find a truth assignment to the variables p and q that satisfies both clauses simultaneously.


3) $p 갿 (춻p 갿 q) 갾 r$ Two clauses, each of which contains two or three variables. The goal is to find a truth assignment to the variables p, q, and r that satisfies both clauses simultaneously.

### 1)
***

```python
for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            if (p or not q) and (not p or q or r) and (not p):
                print(f"Expression 1 is true for p={p}, q={q}, and r={r}.")
```

### 2)
***

```python
for p in [True, False]:
    for q in [True, False]:
        if p or (q and (not p or not q)):
            print(f"Expression 2 is true for p={p} and q={q}.")
```

### 3)
***

```python
for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            if p or (not p or q) and r:
                print(f"Expression 3 is true for p={p}, q={q}, and r={r}.")
```
