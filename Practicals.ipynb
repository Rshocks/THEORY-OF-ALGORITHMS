### Exercise 1 $f(x) = x^2 + 1$
***

```python
# Define the function
def f(x):
    return x**2 + 1

# Create an array of x values
x = np.linspace(-10, 10, 1000)

# Calculate the corresponding y values
y = f(x)

# Plot the function
plt.plot(x, y)

# Set the axis labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x^2 + 1')

# Display the plot
plt.show()
```

### Exercise 2 $g(x) = 3x^6 + x^4 + 1$
***

```python
def g(x):
    return 3*x**6 + x**4 + 1

x = np.linspace(-2, 2, 1000)

y = g(x)

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Plot of g(x) = 3x^6 + x^4 + 1')

plt.show()
```

### Exercise 3 $ x^2 $ and $2^x$ on the same plot
***

```python
# create a range of x-values
x = np.linspace(0, 5, 100)

# calculate y-values for x^2 and 2x
y1 = x**2
y2 = 2**x

# plot the functions on the same graph
plt.plot(x, y1, label='x^2')
plt.plot(x, y2, label='2x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x^2 and 2x')
plt.legend()

plt.show()
```

### Exercise 4 $g(f(x))$
***

```python
def gof(f):
    return 3 * f**6 + f**4 + 1

f = np.linspace(-1, 1, 1000)

gof_values = gof(f)

plt.plot(f, gof_values)

plt.xlabel('f')
plt.ylabel('gof')
plt.title('Plot of gof = 3 * f**6 + f**4 + 1')

plt.show()
```

## Turing Machine
***
A Turing machine is a hypothetical computer that uses a tape with symbols, a read-write head, and a set of rules to perform computations. It can simulate any algorithm and is a theoretical basis for modern computers

```python
# The huring machine tape
tape = " "

# Pos on the tape
pos = 0

# show tape and pos
print(f'{tape}, {pos}')

# moving pos left of tape
def move_L():
    global tape, pos
    if pos > 0:
        pos = pos - 1
    else:
        tape = ' ' + tape
```

```python
move_L()

print(f'{tape}, {pos}')
```

```python
# moving pos right of tape
def move_R():
    global tape, pos
    if pos >= len(tape):
        tape = ' ' + tape
    pos = pos + 1
```

```python
move_R()

print(f'{tape}, {pos}')
```

## Python Function
***
A exercise in python at 22/02/23,     
Given two lists of strings L1 and L2 of the same length, return a list L of indices of L1 and L2 such that L1[i] and L2[j] have the same characters for all positions in the range [0, len(L1)), and such that |L| <= 10.

```python
def correspondence(L1, L2):
    L = []
    for i in range(len(L1)):
        for j in range(len(L2)):
            if i not in L and j not in L and L1[i] == L2[j]:
                L.append(i)
                L.append(j)
                if len(L) >= 10:
                    return L
    return L

L1 = ['010', '001', '111', '000']
L2 = ['010', '101', '011', '000']
L = correspondence(L1, L2)
print(L)
```
